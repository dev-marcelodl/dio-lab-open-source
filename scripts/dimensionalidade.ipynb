```python
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def convert_to_grayscale(image):
    width, height = image.size
    grayscale_image = Image.new("L", (width, height))

    for x in range(width):
        for y in range(height):
            r, g, b = image.getpixel((x, y))
            gray = int(0.3 * r + 0.59 * g + 0.11 * b)
            grayscale_image.putpixel((x, y), gray)

    return grayscale_image
```

```python
def binarize_image(image, threshold=128):
    width, height = image.size
    binary_image = Image.new("1", (width, height))

    for x in range(width):
        for y in range(height):
            gray = image.getpixel((x, y))
            if gray < threshold:
                binary_image.putpixel((x, y), 0)  # Preto
            else:
                binary_image.putpixel((x, y), 1)  # Branco

    return binary_image
```

```python
def plot_images(original, grayscale, binary):
  fig, axs = plt.subplots(1, 3, figsize=(15, 5))
  axs[0].imshow(original)
  axs[0].set_title('Imagem Original')
  axs[0].axis('off')
  axs[1].imshow(grayscale, cmap='gray')
  axs[1].set_title('Imagem em Tons de Cinza')
  axs[1].axis('off')
  axs[2].imshow(binary, cmap='gray')
  axs[2].set_title('Imagem Binarizada')
  axs[2].axis('off')
  plt.show()
```

```python
image = Image.open("lucky.jpg")
grayscale_image = convert_to_grayscale(image)
binary_image = binarize_image(grayscale_image, threshold=100)

plot_images(image, grayscale_image, binary_image)
```
